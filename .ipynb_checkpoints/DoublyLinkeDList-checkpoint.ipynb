```python
class Node:
    
    def __init__(self, _val):
        self.val = _val;
        self.next = None
        self.prev = None
```

```python
class MyLinkedList:

    def __init__(self):
        self.head = None
        self.length = 0
        
        
    def get(self, index):
        if self.head is None:
            return -1

        if index < 0 or index >= self.length:
            return -1
        
        curr = self.head
        for i in range(index):
            curr = curr.next
            
        return curr.val
        
    def addAtTail(self, val):
        newTail = Node(val)
        if self.head is None:
            head = newTail
        else:
            curr = self.head
            while curr.next is not None:
                curr = curr.next
            curr.next = newTail
            newTail.prev = curr
        self.length += 1
        
    def addAtIndex(self, index, val):
        if index < 0 or index > self.length:
            return
        if index == 0:
            self.addAtHead(val)
        elif index == self.length:
            self.addAtTail(val)
        else:
            newNode = Node(val)
            curr = self.head
            for i in range(index - 1):
                curr = curr.next
            curr.next.prev = newNode
            newNode.next = curr.next
            curr.next = newNode
            newNode.prev = curr
            self.length += 1
            
    def toString(self):
        if self.head is None:
            return ""
        curr = self.head
        myStr = ""
        while curr is not None:
            myStr += str(curr.val)
            myStr += " "
            curr = curr.next
            
        return myStr
    
    def addAtHead(self, val):
        newHead = Node(val)
        if self.length == 0:
            self.head = newHead
            self.length += 1
        else:
            newHead.next = self.head
            self.head.prev = newHead
            self.head = newHead
            self.length += 1
            
    def deleteAtIndex(self, index):
        if index < 0 or index >= self.length:
            return
        if index == 0:
            self.head = self.head.next
            self.length -= 1
        elif index == self.length - 1:
            curr = self.head
            while curr.next is not None:
                curr = curr.next
            curr.prev.next = None
            self.length -= 1
        else:
            curr = self.head
            for i in range(index - 1):
                curr = curr.next
            curr.next = curr.next.next
#             if curr.next != None:
            curr.next.prev = curr
            self.length -= 1
```
